## Data Cleaning

```python
df = df_raw.copy()
```

```python
df.drop(columns=["ESG_check"], inplace=True)

for col in ["Industry","Region","CompanyName"]:
    df[col] = df[col].astype("category")

df.info()
```

## Company Selection

```python
eu_com = df[
    (df["Region"] == "Europe") & 
    (df["Industry"] == "Manufacturing") & 
    (df["Year"] == 2025)
].copy()

eu_com["ESG_rank"] = eu_com["ESG_Overall"].rank(ascending=False)
eu_com["CO2_rank"] = eu_com["CarbonEmissions"].rank(ascending=True)

eu_com[["CompanyID", "ESG_Overall", "CarbonEmissions", "Revenue", "ESG_rank", "CO2_rank"]].sort_values("ESG_rank")
```

```python
eu_trend = df[
    (df["Region"] == "Europe") & 
    (df["Industry"] == "Manufacturing") & 
    (df["Year"].isin([2015, 2025]))
].pivot_table(
    index="CompanyID",
    columns="Year",
    values=["ESG_Overall","CarbonEmissions"]
).round(1)

eu_trend["ESG_delta"] = eu_trend["ESG_Overall"][2025] - eu_trend["ESG_Overall"][2015] #trend absolut
eu_trend["CO2_delta"] = (
    (eu_trend["CarbonEmissions"][2025] - eu_trend["CarbonEmissions"][2015]) / eu_trend["CarbonEmissions"][2015] * 100 #prozentual
).round(1)

eu_trend.sort_values("ESG_delta", ascending=False)
```

### Company selection decision:

Company_300 is selected based on a two-dimensional ranking across all 15 European manufacturers. Company_300 shows both a strong reduction in CarbonEmissions as well as a significant increase in ESG Overall scoring.

Among the European manufacturers Company_300 is the only entity showing both an increase in ESG scoring and reduction in CarbonEmissions, which is in line with the CSRDs double materiality principle.
Other top ESG improvers for example Company_224 show rising Emissions which would raise greenwashing concerns under ESRS E1 disclosure requirements.

Company_300 will serve as Lauschke GmbH for this ESG performance analysis.

## Lauschke GmbH dataset

```python
df_l = df[df["CompanyID"] == 300].copy().reset_index(drop=True)

df_eu = df[
    (df["Region"] == "Europe") & 
    (df["Industry"] == "Manufacturing")
].copy()

df_l
```
